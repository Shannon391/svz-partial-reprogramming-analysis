```r
## Celltype subset for each Age_Treatment Group within a Cell-Type Group
Celltype_Subset <- function(seuListInput, cellType){
    celltype_subset_seu_list <- list()
    young_ctrl_subset_seu_list <- list()
    old_ctrl_subset_seu_list <- list()
    old_OSKM_subset_seu_list <- list()
    for(i in 1:length(seuListInput)){
        for(j in 1:length(seuListInput[[i]])){
            if(i == 1){
                if(cellType %in% levels(factor(seuListInput[[1]][[j]]@meta.data$Celltype))){
                    young_ctrl_subset_seu_list <- c(young_ctrl_subset_seu_list, subset(seuListInput[[i]][[j]], Celltype == cellType))
                }
            }
            else if(i == 2){
                if(cellType %in% levels(factor(seuListInput[[i]][[j]]@meta.data$Celltype))){
                    old_ctrl_subset_seu_list <- c(old_ctrl_subset_seu_list, subset(seuListInput[[i]][[j]], Celltype == cellType))
                }
            }
            else if(i == 3){
                if(cellType %in% levels(factor(seuListInput[[i]][[j]]@meta.data$Celltype))){
                    old_OSKM_subset_seu_list <- c(old_OSKM_subset_seu_list, subset(seuListInput[[i]][[j]], Celltype == cellType))
                }
            }   
        }
    }
    
    celltype_subset_seu_list$young_ctrl <- young_ctrl_subset_seu_list
    celltype_subset_seu_list$old_ctrl <- old_ctrl_subset_seu_list
    celltype_subset_seu_list$old_OSKM <- old_OSKM_subset_seu_list

    celltype_subset_seu_list
}
```

```r
# Makes Transcriptional_Drift Function with a list of lists of seurat objects as input (old seurat object 
# of one age_treatment group first, and young control seurat object second)
Transcriptional_Drift <- function(seuInput){

    # Stores vector of old (control or treatment) counts means (length = # genes * 1)
    old_means_df <- as.data.frame(lapply(X = seuInput[[2]], FUN = function(oldSeuInput){
            apply(X = oldSeuInput@assays[["SCT"]]@data, MARGIN = 1, FUN = mean)
        }))
    
    # Stores vectors of young controls' counts means (each has length = # genes * 1) in young_ctrl_mean_df
    young_ctrl_mean_df <- as.data.frame(lapply(X = seuInput[[1]], FUN = function(youngSeuInput){
            apply(X = youngSeuInput@assays[["SCT"]]@data, MARGIN = 1, FUN = mean)
        }))
    
    # Computes means of each gene (across rows in young_ctrl_mean) and stores single vector in avg_young_ctrl_mean
    avg_young_ctrl_mean <- apply(X = young_ctrl_mean_df, MARGIN = 1, FUN = mean)

    # Calculates transcriptional drift values (returns a single vector of length = # genes * 1)
    apply(X = old_means_df, MARGIN = 2, FUN = function(oldMeansInput){
        oldMeansInput/avg_young_ctrl_mean
        })
}
```

```r
# Drift Variance function that takes a dataframe of log(transcriptional drift) data of an age_treatment group of 
# individual mouse samples

Drift_Variance <- function(tdGroupInput){
    tdGroupInput[is.na(tdGroupInput)] <- 0
    tdGroupInput[is.infinite(tdGroupInput)] <- 0
    log_tn <- apply(tdGroupInput, MARGIN = c(1,2), FUN = log)
    log_tn[is.na(log_tn)] <- 0
    log_tn[is.infinite(log_tn)] <- 0
    dv <- apply(X = log_tn, MARGIN = 2, FUN = var)
    dv
}
```
